```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("cleaned_austin_weather.csv")
data
```

```python
X = data.drop("PrecipitationSumInches", axis=1)
```

```python
Y = data["PrecipitationSumInches"]
```

```python
Y = Y.values.reshape(-1, 1)
```

```python
day_index = 798
days = [i for i in range(Y.size)]
```

```python
clf = LinearRegression()
clf.fit(X, Y)
```

```python
print(type(plt.title))
```

```python
print("The precipitation trend graph")
plt.scatter(days, Y, color="aqua", label="Precipitation")
plt.scatter(days[day_index], Y[day_index], color="red", label="Selected Day")
plt.title("Precipitation Level Over Days")        
plt.xlabel("Days")                               
plt.ylabel("Precipitation in Inches")               
plt.legend()
plt.show()
x_vis = X.filter(["TempAvgF", "DewPointAvgF", "HumidityAvgPercent","SeaLevelPressureAvgInches", "VisibilityAvgMiles", "WindAvgMPH"])
```

```python
print("The precipitation trend graph VS attributes trend graph: ")

for i in range(x_vis.columns.size):
    plt.subplot(3,2,i+1)
    plt.scatter(days, x_vis[x_vis.columns.values[i][:100]])
    plt.scatter(days[day_index],x_vis[x_vis.columns[i]][day_index])
plt.show()
               
```
